## Merge

```python
# Merge SHP file and CSV file
world_map = poligonos.merge(latest_data, on = "Code", how = "right")
```

```python
world_map
```

```python
world_map = world_map.rename(columns={"total suicides/100k pop":"SuicideRate"})
world_map.head()
```

```python
type(world_map)
```

```python
world_map["geometry"]
```

```python
world_map["Code"].value_counts()
```

```python
list(world_map.columns)
```

```python
world_map = world_map[[
    "Country",
    "Code",
    "year",
    "suicides_no",
    "population",
    "SuicideRate",
    "gdp_for_year ($)",
    "gdp_per_capita ($)",
    "Region",
    "2020 Rank",
    "2020 Score",
    "Location",
    "NAME",
    "LON",
    "LAT",
    "geometry"
]]
```

```python
# transfomracion de poligonos
world_map.to_crs("EPSG:4326")
world_map.head()
```

```python
world_map["geometry"] = world_map["geometry"].simplify(tolerance = 0.001, preserve_topology = True)
```

```python
# get geojson file for world map
world_map.to_file("static/data/world_map.geojson", driver='GeoJSON')
```
